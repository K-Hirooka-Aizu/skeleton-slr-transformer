```python
import os
import cv2
```

```python
VIDEO_DATA_DIR = "/mnt/content/DATASET/AUTSL/video/test"

target_video_path = []
for root, dirs, files in os.walk(VIDEO_DATA_DIR):
    for file in files:
        if file.endswith(".mp4") and "ipynb_chenckpoint" not in os.path.join(root,file):
            target_video_path.append(os.path.join(root,file))
```

```python
def process(path,model):
    cap = cv2.VideoCapture(path)

    if not cap.isOpened():
        return


    n = 0

    while True:
        ret, frame = cap.read()
        if ret:
            keypoint, H, W = model(frame)
            print(keypoint.shape)
            n += 1
        else:
            return


model = Model()
for path in target_video_path:
    process(path,model)
```
